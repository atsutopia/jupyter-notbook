```python
import pandas as pd

df = pd.read_csv('scount_2020_06_15-v2.csv')
df.created_at = pd.to_datetime(df.created_at, utc=True)
df.index = pd.DatetimeIndex(df.created_at, name="created_at")
df.index = df.index.tz_convert('Asia/Tokyo')

df.
df.head()
```

```python
from datetime import time

print(len(df))

time1 = df.between_time(time(9, 0),time(12, 0))
print('time1', len(time1),  (time1['is_interested'] == 1.0).sum())

time2 = df.between_time(time(12, 0),time(15, 0))
print('time2', len(time2),  (time2['is_interested'] == 1.0).sum())

time3 = df.between_time(time(15, 0),time(18, 0))
print('time3', len(time3),  (time3['is_interested'] == 1.0).sum())

time4 = df.between_time(time(18, 0),time(21, 0))
print('time4', len(time4),  (time4['is_interested'] == 1.0).sum())

time5 = df.between_time(time(21, 0),time(0, 0))
print('time5', len(time5),  (time5['is_interested'] == 1.0).sum())

time6 = df.between_time(time(0, 0),time(9, 0))
print('time6', len(time6),  (time6['is_interested'] == 1.0).sum())

```
